2D Array Factor

```python
from math import *
import numpy as np
from IPython import display
import matplotlib.pyplot as plt

th_res = 0.001
theta0 = np.arange(th_res,2*pi,th_res)
print(theta0)

N = 5
AF = [] # Initialize
f = 28 * (10**9)
c = 3 * (10**8)
l = c/f
d = 0.5 * l
k = 2*pi/l

for i in range(0,len(theta0)):
  phi = k * d * cos(i * th_res)
  f_phi = sin(N*phi/2)/(N*sin(phi/2))
  AF.append(abs(f_phi))

display.clear_output(wait=True)
display.display(plt.gcf())
plt.figure().clear()
plt.axes(polar=True)
plt.plot(theta0,AF)
plt.xlim(0,pi)
```

2D Beam Shift

```python
from math import *
import numpy as np
from IPython import display
import matplotlib.pyplot as plt

th = 15 * (pi/180) # Phase Shift
th_res = 0.001 #1 * (pi/180)
theta0 = np.arange(th_res,2*pi,th_res)
print(theta0)

N = 5
AF = [] # Initialize
f = 28 * (10**9)
c = 3 * (10**8)
l = c/f
d = 0.5 * l
k = 2*pi/l

for i in range(0,len(theta0)):
  f_phi = 0
  phi = k * d * cos(i * th_res - th)
  for j in range(0, N):
    f_phi = f_phi + complex(cos(j * phi),sin(j * phi))
  AF.append(abs(f_phi))

display.clear_output(wait=True)
display.display(plt.gcf())
plt.figure().clear()
plt.axes(polar=True)
plt.plot(theta0,AF)
plt.xlim(0,pi)
```
